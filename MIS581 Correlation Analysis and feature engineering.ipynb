```python
# significance of correlation using p value 
# listing all r and p values for the dependent variable is_cancelled and predictor variables

from scipy.stats import pearsonr
import pandas as pd
df = bookings._get_numeric_data()
dfcols = pd.DataFrame(columns=df.columns)
c='is_canceled'
#print(dfcols)
for r in dfcols:
        coeff = pearsonr(df[r], df[c])[0]
        pvalue = pearsonr(df[r], df[c])[1]
        print(r,',',c,',',coeff,',',pvalue)
```
